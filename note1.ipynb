```python
import geopandas as gpd
import matplotlib.pyplot as plt
gdf = gpd.read_file('data/kuching_output.geojson')
print(gdf.head())
```

```python
gdf.plot()
plt.title('Kuching')
plt.show()
```

```python
gdf.info()
```

```python
len(gdf['name'])
```

```python
gdf_cop = gdf.copy()
```

```python
num_nans = gdf_cop.isnull().sum(axis=1)
threshold = 5
gdf_cop_drop = gdf_cop[num_nans <= threshold]
```

```python
print(gdf_cop_drop.head())
```

```python
gdf_cop_drop.plot()
plt.title('Kuching')
plt.show()
```
